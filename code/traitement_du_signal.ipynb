````python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import scipy
````

````python
# liste vide pour stocker les valeurs du signal brut
signal_brut = []

# ouverture du fichier en lecture
fichier = open('samples/1551279158_nico_contractions_courtes_moyennes.txt', 'r')

# lecture des données
nb_lignes = 0 # permets de compter les lignes
for ligne in fichier: # chaque ligne est un objet string
    if ligne.startswith('#'): # on affiche les lignes de commentaires
        print(ligne)
    else:
        signal_brut.append(int(ligne))
        nb_lignes += 1

# fermeture du fichier
fichier.close()

# graphique du signal brut
plt.plot(signal_brut)
````
